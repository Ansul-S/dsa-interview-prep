# Arrays

## Types of Arrays

### 1. Homogeneous Arrays (Homo)
- Store **elements of the same data type**
- Example: **C Language Arrays**
- All elements must be uniform  
- Example:

```c
int arr[] = {1, 2, 3, 4};
```

---

### 2. Heterogeneous Arrays (Hetero)
- Can store **different data types**
- Common in **Python** (especially lists)

Example:

```python
arr = [10, "Hello", 3.14, True]
```

---

## Definition of an Array

An array is:

- A **collection of similar elements**  
- Stored in **contiguous memory locations**  
- Allows efficient access using **indexing**

> **Note:**  
> Traditional arrays usually contain similar data types,  
> but Python structures can also support mixed data types.

---

## Contiguous Memory Allocation

Arrays store elements in **continuous (adjacent) memory locations**.

Example:

| Index | Value | Memory Location |
|-------|-------|----------------|
| 0 | 10 | 1000 |
| 1 | 20 | 1004 |
| 2 | 30 | 1008 |

This enables:

- Faster access  
- Efficient traversal  
- Better memory management

---

## Accessing Arrays in Python

### 1. Using Python `array` Module

```python
import array as arr

a = arr.array('i', [10, 20, 30])

print(a[0])   # Access first element
```

---

### 2. Using NumPy

```python
import numpy as np

a = np.array([10, 20, 30])

print(a[1])   # Access second element
```

---

## Summary

| Feature | Homogeneous | Heterogeneous |
|---------|------------|---------------|
| Same Data Type | Yes | No |
| C Arrays | Yes | No |
| Python Lists | No | Yes |
| NumPy Arrays | Yes | No |


# Array Module

## Python Array Module (`array`)

## Importing the Array Module

Python provides the built-in `array` module for creating **homogeneous arrays** (elements must be of the same type).

```python
import array as arr
```

You can also import directly:

```python
from array import *
```

---

## Creating an Array

### Syntax

```python
array.array(typecode, elements)
```

### Example

```python
import array as arr

a = arr.array('i', [10, 20, 30, 40])

print(a)
```

Output:

```python
array('i', [10, 20, 30, 40])
```

# Type Codes in Python Array

| Type Code | Data Type | Size (Bytes) | Description |
|----------|-----------|--------------|-------------|
| `'b'` | int | 1 | Signed integer |
| `'B'` | int | 1 | Unsigned integer |
| `'u'` | Unicode | 2 | Unicode character |
| `'h'` | int | 2 | Signed short integer |
| `'H'` | int | 2 | Unsigned short integer |
| `'i'` | int | 2 or 4 | Signed integer |
| `'I'` | int | 2 or 4 | Unsigned integer |
| `'l'` | int | 4 | Signed long integer |
| `'L'` | int | 4 | Unsigned long integer |
| `'f'` | float | 4 | Floating point |
| `'d'` | float | 8 | Double precision float |

# Common Array Functions

## 1. append()

Adds element at end.

```python
a.append(50)
```

---

## 2. insert()

Insert at specific position.

```python
a.insert(1,100)
```

---

## 3. remove()

Removes element.

```python
a.remove(20)
```

---

## 4. pop()

Removes element by index.

```python
a.pop(2)
```

---

## 5. index()

Finds position.

```python
a.index(30)
```

---

## 6. reverse()

Reverses array.

```python
a.reverse()
```

---

## 7. extend()

Adds multiple elements.

```python
b = arr.array('i', [60,70])

a.extend(b)
```

---

## 8. buffer_info()

Shows memory address and length.

```python
a.buffer_info()
```

Example Output:

```python
(140673456789456, 5)
```

- First value → memory address  
- Second value → number of elements

---

## 9. count()

Counts occurrences.

```python
a.count(10)
```

---

## 10. tolist()

Convert array to list.

```python
a.tolist()
```

---

## 11. fromlist()

Convert list into array.

```python
a.fromlist([100,200])
```

---
