### Q2: Second Largest

Given:

arr = [10, 5, 8, 20, 15]

Find the second largest element.

Constraint:
- Try solving in one pass.

Questions:
1. Brute force?
2. One-pass solution?

```python
# Brute Force
arr = [10, 5, 8, 20, 15]

# find largest
first = arr[0]
for x in arr:
    if x > first:
        first = x

# find second largest
second = float('-inf')
for x in arr:
    if x != first and x > second:
        second = x

print(second)
```

```python
# One-pass solution
arr = [10, 5, 8, 20, 15]

first = float('-inf')
second = float('-inf')

for x in arr:
    if x > first:
        second = first
        first = x
    elif x < first and x > second:
        second = x

print(second)
```

```python
# Find second largest in ONE PASS

arr = [10, 5, 8, 20, 15]

# We use -infinity so that ANY number in the array will be bigger initially
first = float('-inf')    # largest so far
second = float('-inf')   # second largest so far

for x in arr:
    # Case 1: Found a new largest number
    if x > first:
        second = first   # old largest becomes second largest
        first = x        # update largest

    # Case 2: Not largest, but can be second largest
    elif x < first and x > second:
        second = x

# After one full pass, 'second' holds the answer
print("Second largest:", second)
```

```python
# Optimal Approch
def second_largest(arr):
    if len(arr) < 2: # if array has 0 or 1 elements,there can't be a second largest 
        return None
    
    max_val = arr[0] # the largest number seen so far. Start with first element
    second_max = float('-inf') # the second largest seen so far. Start with negative infinity, which basically means "nothing found yet" but Python can still compare numbers against it
# Think of it like two empty podium slots 1st and 2nd place

    for x in arr:
        if x > max_val:
            second_max = max_val # old #1 drops to #2
            max_val = x # new #1 takes the throne
        elif x > second_max and x != max_val:
            second_max = x # x != max_val handles duplicates. If array is [9, 9, 5], we don't want 9 to be both #1 and #2
    
    return second_max if second_max != float('-inf') else None #If second_max was never updated from -inf, no second largest exists — return None
print(second_largest(arr))
```

### Topic: Reverse Array
* Before: [1,2,3,4,5]
* After:  [5,4,3,2,1]

```python
def reverse_arr(arr):
    left = 0
    right = len(arr) - 1

    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        left = left + 1
        right = right - 1

    return arr
print(reverse_arr([1, 2, 3, 4, 5]))
```

```python
# Problem:        Reverse an array in-place
# Pattern:        Two pointers — left and right walking inward
# Key Trick:      Swap arr[left] and arr[right], then move both
#                pointers inward until they meet
# Mistake I made: Confused pointer movement with value swapping
#                Confused while condition — use left < right
# When to use:    Any time you need to mirror/reverse in-place
#                Palindrome checks, in-place rotations
```

```python
def move_zeroes(arr):
    slow = 0
    
    for fast in range(len(arr)):
        if arr[fast] != 0: # what are we looking for?
            arr[slow], arr[fast] = arr[fast], arr[slow]
            slow = slow + 1 # move slow foward
    
    return arr
```

```python
# Remove Duplicates
# Input:  [1, 1, 2, 3, 3, 4]
# Output: [1, 2, 3, 4]
```

```python

def remove_duplicates(arr):
    if not arr:
        return []
    
    slow = 0
    
    for fast in range(1, len(arr)):  # start from 1, first element always kept
        if arr[fast] != arr[slow]:   # different from slow's element?
            slow = slow + 1
            arr[slow] = arr[fast]    # place it at slow's position
    
    return arr[:slow + 1]            # return only unique portion

print(remove_duplicates([1, 1, 2, 3, 3, 4]))  
print(remove_duplicates([1, 1, 1, 1]))         
print(remove_duplicates([1, 2, 3, 4]))        
print(remove_duplicates([]))     
```

```python
# Problem:        Remove duplicates from sorted array
# Pattern:        Two pointers — slow and fast, both from left
# Key Trick:      Copy don't swap — arr[slow] = arr[fast]
#                Condition: arr[fast] != arr[slow]
#                Works because array is sorted — duplicates
#                are always neighbours
# Mistake I made: Compared against 0 (Move Zeroes habit)
#                Used swap syntax instead of copy
# When to use:    Removing consecutive duplicates
#                Sorted array cleanup problems
```

### Problem — "Clean and Analyze"
You are given a raw, unsorted array of integers collected from a biology experiment. The data is messy. Your job is to clean and analyze it in exactly this order:

* Step 1 → Move all zeroes to the end
* Step 2 → Remove duplicates from the 
* non-zero portion
* Step 3 → Find the maximum value
* Step 4 → Find the second largest value
* Step 5 → Reverse the cleaned array
* Step 6 → Return all results

#### Array Processing Results

**Input:**  
`[0, 3, 1, 3, 0, 7, 2, 7, 1, 0]`

---

**After moving zeroes:**  
`[3, 1, 3, 7, 2, 7, 1, 0, 0, 0]`

**After removing duplicates:**  
`[3, 1, 7, 2]`

**Maximum:**  
`7`

**Second Largest:**  
`3`

**After reversing:**  
`[2, 7, 1, 3]`

#### Rules

- Solve each step using what you learned — **no built-ins**
- Write a **separate function** for each step
- Call them **one by one in order**
- Handle **empty array** at each step



```python
arr = [0, 3, 1, 3, 0, 7, 2, 7, 1, 0]
def move_zeroes(arr):
    if not arr:
        return []
    
    slow = 0
    
    for fast in range(len(arr)):
        if arr[fast] != 0:
            arr[slow],arr[fast] = arr[fast], arr[slow]
            slow += 1
    return arr


def remove_duplicates_unsorted(arr):
    seen = set()
    result = []
    for x in arr:
        if x != 0 and x not in seen:  # skip zeroes too
            seen.add(x)
            result.append(x)
    return result

def find_maximum(arr):
    if not arr:
        return None
    
    max_val = arr[0]
    for x in arr[1:]:
        if x > max_val:
            max_val = x
    return max_val

def find_second_largest(arr):
    max_val = arr[0]
    sec_val = float('-inf')

    for x in arr:
        if x > max_val:
            sec_val = max_val
            max_val = x
        elif x > sec_val and x < max_val:
            sec_val = x
    return sec_val if sec_val != float('-inf') else None

def reverse_array(arr):
    left = 0
    right = len(arr)-1

    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        left = left + 1
        right = right - 1
    return arr

def clean_and_analyze(arr):
    print("Original:                ", arr)

    arr = move_zeroes(arr.copy())
    print("After moving zeroes:     ", arr)

    arr = remove_duplicates_unsorted(arr)
    print("After removing duplicates:", arr)

    max_val = find_maximum(arr)
    print("Maximum:                ", max_val)

    sec_val = find_second_largest(arr)
    print("Second Largest:         ", sec_val)

    arr = reverse_array(arr.copy())
    print("After reversing:        ", arr)

    return arr, max_val, sec_val

clean_and_analyze(arr)
```

## 🧩 Problem — "Segment and Reforge"

You are given an unsorted integer array from a corrupted sensor log. The data contains **negative values, duplicates, and noise**. You must process it step by step:

---

### ⚔️ Steps

**Step 1 → Move all negative numbers to the beginning**  
(Maintain relative order of non-negative numbers)

**Step 2 → Remove duplicates from the non-negative portion only**

**Step 3 → Find the minimum value from the cleaned non-negative portion**

**Step 4 → Find the third largest value from the cleaned non-negative portion**

**Step 5 → Rotate the cleaned non-negative array to the right by 2 positions**

**Step 6 → Return all results**

---

### 🧪 Input
[-3, 5, 1, -2, 5, 7, 1, 0, -1, 7]


---

### 🔍 Expected Processing

**After moving negatives:**  
[-3, -2, -1, 5, 1, 5, 7, 1, 0, 7]


**After removing duplicates (non-negative only):**  
[-3, -2, -1, 5, 1, 7, 0]


**Minimum (non-negative portion):**  
0


**Third Largest (non-negative portion):**  
1


**After rotating right by 2:**  
[-3, -2, -1, 7, 0, 5, 1]


---

### ⚙️ Rules

- Do **not** use built-in shortcuts (`sort`, `set`, etc.)
- Write a **separate function** for each step
- Call functions **one by one in order**
- Maintain **in-place logic where possible**
- Clearly define the **non-negative portion** (after Step 1)
- Handle edge cases:
  - Less than 3 unique values
  - All negatives
  - Empty array



```python
arr = [-3, 5, 1, -2, 5, 7, 1, 0, -1, 7]

def neg_beg(arr):
    j = 0
    for i in range(len(arr)):
        if arr[i] < 0:
            arr[i], arr[j] = arr[j], arr[i]
            j += 1
    return arr, j
print(neg_beg(arr))
```

```python
def remove_duplicates(arr, start, end):
    #1. Edge case check
    # If the segment is empty or has 1 element → nothing to do
    # Example:
    # start = 3, end = 3 → empty
    # start = 3, end = 4 → one element
    # So we just return.
    if start >= end:
        return start
    #2. Initialize write pointer
    # Think of write as:
    # “Where should I place the next unique value?”
    # Initially, it points to the start of the segment. 
    write = start
    
    #3. Loop through the segment
    # i scans every element in the segment
    # this is your reader pointer
    for i in range(start, end):
        #4. we'll try to prove if its duplicate
        duplicate = False
         
        #5. Check if already seen
        # this is the core logic
        # we only check already accepted unique elements
        # that range is [start, write]
        # why? because everything before write is already cleaned and unique
        for k in range(start, write):
            if arr[k] == arr[i]:
                duplicate = True
                break

        #6. If not duplicate -> keep it
        # Two things happen:
        # 1. we place the value at the write position
        # 2. Move write forward
        # so the array gradually becomes:
        # [unique values | old values]
        if not duplicate:
            arr[write] = arr[i]
            write += 1
    return write
```
