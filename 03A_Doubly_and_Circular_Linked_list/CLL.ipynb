###  Circular Linked List

A **circular linked list** is a linked list where the **last node points back to the first node (head)** instead of `None`.

---

###  Key Difference

- Normal Linked List:  
  `last.next = None`

- Circular Linked List:  
  `last.next = head` 

---

###  Example

10 → 20 → 30 → 40  
↑               ↓  
└───────────────┘  

- Head = 10  
- Last node = 40  
- 40 points back to 10 (head)

---

###  Traversal

Since there is no `None`, we stop when we reach the head again:

```python
temp = self.head

while temp.next != self.head:
    print(temp.data)
    temp = temp.next

print(temp.data)  # print last node

###  Doubly Circular Linked List

A **doubly circular linked list** is a linked list where:

- Each node has **two pointers**:
  - `next` → points to next node
  - `prev` → points to previous node
- The list is **circular**:
  - Last node connects back to the first
  - First node connects back to the last

---

###  Key Adjustments

To make a doubly linked list **circular**, we update two pointers:

1. **Last node → first node**
   ```python
   last.next = head

2. **First node → last node**
    ```python
    head.prev = last
    
Example:
NULL ← 10 ⇄ 20 ⇄ 30 ⇄ 40 → NULL   (Normal DLL)

10 ⇄ 20 ⇄ 30 ⇄ 40
↑                 ↓
└─────────────────┘   (Doubly Circular DLL)

Head = 10
Last = 40
40.next → 10
10.prev → 40

### Traversal (Forward)

Since there is no None, stop when we reach the head again:

```python
temp = head

while temp.next != head:
    print(temp.data)
    temp = temp.next

print(temp.data)  # last node
```
### Traversal (Backward)

```python
temp = head.prev  # start from last node

while temp.prev != head.prev:
    print(temp.data)
    temp = temp.prev

print(temp.data)  # first node
```
###  Why Use Doubly Circular Linked List?
* Easy forward + backward traversal
* Efficient for navigation systems (undo/redo)
* Used in circular buffers / schedulers


````python
class Node():
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def append(self,data):
        new_node = Node(data)

        # Situation 1 - empty list
        if self.head is None:
            new_node.next = new_node # points to itself
            self.head = new_node
            return

        # Situation 2 - travel to last node
        current = self.head
        while current.next != self.head: # last node points to head
            current = current.next
        
        current.next = new_node # connect last to new
        new_node.next = self.head

    def traverse(self):
        if self.head is None:
            return
        current = self.head
        while True:
            print(current.data)
            current = current.next
            if current == self.head:     # looped back to start?
                break

cll = CircularLinkedList()
cll.append(1)
cll.append(2)
cll.append(3)
cll.append(4)

print("Traverse:")
cll.traverse()
````
