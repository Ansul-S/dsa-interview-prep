# Doubly Linked List

Each node has **3 parts**:

---

##  Example with 2 Nodes

###  First Node (Head)
- `head` points to this node
Prev = None | Data = 10 | Next = address of 2nd node

Meaning:
- No previous node → `Prev = None`
- Stores value `10`
- Points forward to the second node

---

### Second Node (Temp)
- `temp` points to this node
Prev = address of 1st node | Data = 20 | Next = None

 Meaning:
- Points back to first node
- Stores value `20`
- No next node → `Next = None`

---

##  Visual Structure
None ← [10] ⇄ [20] → None
↑ ↑
head temp

---

##  How to Access Data

### From `head` (1st node)
- `head.prev` → None
- `head.data` → 10
- `head.next` → 2nd node

### From `temp` (2nd node)
- `temp.prev` → 1st node
- `temp.data` → 20
- `temp.next` → None

---

##  Key Idea

- Each node connects **both ways**:
  - `prev` → goes backward
  - `next` → goes forward
- First node → `prev = None`
- Last node → `next = None`


## Adding a 3rd Node (Data = 30)

###  Current List
None ← [10] ⇄ [20] → None
↑ ↑
head temp


---

###  Goal
Add a new node with value **30** after the 2nd node

---

##  Step 1: Reach the Last Node

Use a pointer `t` and move it forward:

t = head
while (t.next != None):
t = t.next

Now:
t is pointing to [20] (last node)
##  Step 2: Create New Node

---

##  Step 3: Link the Nodes

Connect the new node with the last node:
t.next = newNode
newNode.prev = t
newNode.next = None

---

##  Final Structure
None ← [10] ⇄ [20] ⇄ [30] → None
↑ ↑ ↑
head temp t

---

##  Key Idea (Super Important)

- Move `t` to the **last node**
- Then:
  - Last node’s `next` → new node
  - New node’s `prev` → last node
- New node becomes the new **end of the list**

```python
# Insert at the End

class Node:
    def __init__(self, value=None):
        self.data = value
        self.next = None
        self.prev = None


class DoublyLL:
    def __init__(self):
        self.head = None

    def insertAtEnd(self, value):
        temp = Node(value)
        if self.head is None:
            self.head = temp
            return

        t = self.head
        while t.next is not None:
            t = t.next
        
        t.next = temp
        temp.prev = t

    def printDLL(self):
        t1 = self.head
        while t1 is not None:
            if t1.next is not None:
                print(t1.data, end = ' <-> ')
            else:
                print(t1.data, end = ' ')
            t1 = t1.next


obj = DoublyLL()
obj.insertAtEnd(10)
obj.insertAtEnd(20)
obj.insertAtEnd(30)
obj.insertAtEnd(40)
obj.printDLL()
```
