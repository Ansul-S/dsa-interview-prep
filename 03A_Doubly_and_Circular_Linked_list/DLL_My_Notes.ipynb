```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        current = self.head
        while current.next != None:
            current = current.next
        current.next = new_node     # forward link
        new_node.prev = current         # backward link ← new
    
    def prepend(self, data):
        new_node = Node(data)
        new_node.next = self.head        # connect forward

        if self.head is not None:
            self.head.prev = new_node    # connect backward

        self.head = new_node             # redirect head

    def delete(self, data):
        if self.head is None:
            return

    # Situation 2 — delete head
        if self.head.data == data:
            self.head = self.head.next
            if self.head is not None:
                self.head.prev = None        # new head has no previous
            return

    # Situation 3 — find target node
        current = self.head
        while current != None:
            if current.data == data:
                current.prev.next = current.next     # tell prev to skip target
                if current.next is not None:
                    current.next.prev = current.prev # tell next to skip back
                return
            current = current.next

        print(f'{data} not found')

    def traverse_forward(self):
        current = self.head
        while current != None:
            print(current.data)
            current = current.next

    def traverse_backward(self):
        # first go to last node
        current = self.head
        while current.next != None:
            current = current.next
        # then walk backwards
        while current != None:
            print(current.data)
            current = current.prev   # which pointer goes backward?

dll = DoublyLinkedList()
dll.append(1)
dll.append(2)
dll.append(3)
dll.append(4)

# Test 1 — delete middle node
dll.delete(2)
print("After deleting 2:")
dll.traverse_forward()

# Test 2 — delete head
dll.delete(1)
print("After deleting head 1:")
dll.traverse_forward()

# Test 3 — delete tail
dll.delete(4)
print("After deleting tail 4:")
dll.traverse_forward()

# Test 4 — not found
dll.delete(9)
```
