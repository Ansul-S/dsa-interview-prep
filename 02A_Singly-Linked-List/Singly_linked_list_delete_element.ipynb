### Singly Linked List
* Delete element

```python
class Node:
    def __init__(self, info, next=None): # info , next are the parameters, so next ka by default parameter set kr diya as None
        self.data = info # We set data as info, info ko maine data name kiya hai info bhi rakh sakta tha  
        self.next = next


class SinglyLinkedList:
    def __init__(self, head=None): # Jitne Linked list banaye utne head honge
        self.head = head

# In linked List we learn insertion
# Insertion at the Beginning, Middle, End
# Insertion at End

    
    def insertAtEnd(self, value):
        temp = Node(value) # isse hamara node bangaya jiske data part main 30 chalagaya or next main None, Now we need to make a pointer which we call temp, Head is pointing 1st location and temp is pointing our end, so we need to create an assistent as t1 which is pointing first location

        if self.head is not None:
            t1 = self.head
            while t1.next is not None:
                t1 = t1.next
            t1.next = temp
            # Now we need to link our temp node and t1 to link. we need to assign temp node address to t1 next
        else:
            self.head = temp
    
    def insertAtBeg(self, value):
        temp = Node(value)
        temp.next = self.head
        self.head = temp

    def insertAtMid(self, value, x): # This Method also help to search the element
        temp = Node(value)
        t1 = self.head

        while t1 is not None:
            if(t1.data == x):
                temp.next = t1.next
                t1.next = temp
                return
            t1 = t1.next
        print("Value to insert after not found")


    def deleteLL(self, value): # This can delete mid and edge cases both
        if self.head is None:
            print("List is empty")
            return

        t1 = self.head
        prev = None

        while t1:
            if t1.data == value:
                if prev:
                    prev.next = t1.next
                else:
                    self.head = t1.next
                return
            prev = t1
            t1 = t1.next
        print("Value not found")
    
    def printLL(self):
        t1 = self.head
        while t1 is not None:
            print(t1.data)
            t1 = t1.next


# usage
obj = SinglyLinkedList()
obj.insertAtEnd(10)
obj.insertAtEnd(20)
obj.insertAtEnd(30)
obj.insertAtBeg(5)
obj.insertAtMid(40, 20)
obj.deleteLL(30)
obj.printLL()
```
