## Interview Problem 1 — Reverse a Linked List

### The Problem
Given a linked list, reverse it **in place**.

**Input:**  
`head → [1] → [2] → [3] → [4] → [5] → None`

**Output:**  
`head → [5] → [4] → [3] → [2] → [1] → None`

````python
class Node():
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
            return

        current = self.head
        while current.next != None:
            current = current.next

        current.next = new_node


    def reverse(self):
        prev = None
        current = self.head

        while current != None:
            next = current.next
            current.next = prev
            prev = current
            current = next
        self.head = prev

    def traverse(self):
        current = self.head
        while current != None:
            print(current.data)
            current = current.next

ll = LinkedList()
ll.append(1)
ll.append(2)
ll.append(3)
ll.append(4)
ll.append(5)

print("Before:")
ll.traverse()

ll.reverse()

print("After:")
ll.traverse()
````

## Problem: Find Middle of a Linked List

### Problem Statement
Given a linked list, find the **middle node** and return its value.

---

### Examples

**Example 1 (Odd Length):**
```text
Input:  head → [1] → [2] → [3] → [4] → [5] → None
Output: 3
```
**Example 2 (Even Length):**
```text
Input:  head → [1] → [2] → [3] → [4] → None
Output: 2 (for even length, return second middle)
```

````python
class Node():
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
            return

        current = self.head
        while current.next != None:
            current = current.next

        current.next = new_node

    def find_middle(self):
        slow = self.head
        fast = self.head

        while fast != None and fast.next != None:
            slow = slow.next
            fast = fast.next.next

        return slow.data

    def traverse(self):
        current = self.head
        while current != None:
            print(current.data)
            current = current.next

ll = LinkedList()
ll.append(1)
ll.append(2)
ll.append(3)
ll.append(4)
ll.append(5)

print("MIDDLE TERM:")
print(ll.find_middle())

ll2 = LinkedList()
ll2.append(1)
ll2.append(2)
ll2.append(3)
ll2.append(4)
print(ll2.find_middle()) 
````

###  Problem: Find Middle Node of a Linked List

**Pattern:**
Slow and fast pointer (Floyd's algorithm)

**Key Trick:**

* Fast pointer moves at **2× speed** of slow
* When fast reaches the end → slow is at the middle

**Stop Condition:**

* `fast != None and fast.next != None`
* Both conditions are required — **order matters**
* `fast.next` is checked *after* `fast` → prevents crash

**Behavior:**

* **Odd length:** returns exact middle
* **Even length:** returns second middle

**Approaches:**

* Naive: two passes → `O(2n)`
* Optimized: one pass → `O(n)` (preferred in interviews)

**Complexity:**

* Time: `O(n)`
* Space: `O(1)`


### Interview Problem 3 — Detect a Cycle

**The Problem:**
Given a linked list, return `True` if there is a cycle, `False` if not.

**No Cycle Example:**

```
head → [1] → [2] → [3] → [4] → None
```

→ `False`

**Cycle Example:**

```
head → [1] → [2] → [3] → [4]
                    ↑      |
                    └──────┘
```

→ `True`

Here, node `[4]` points back to `[3]`, creating an infinite loop.


````python
class Node():
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        current = self.head
        while current.next != None:
            current = current.next
        current.next = new_node

    def detect_cycle(self):
        slow = self.head
        fast = self.head
        while fast != None and fast.next != None:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                return True
        return False

# Test 1 — no cycle
ll1 = LinkedList()
ll1.append(1)
ll1.append(2)
ll1.append(3)
ll1.append(4)
print(ll1.detect_cycle())   # False

# Test 2 — with cycle
ll2 = LinkedList()
ll2.append(1)
ll2.append(2)
ll2.append(3)
ll2.append(4)
# [4] points back to [3] — creating cycle
ll2.head.next.next.next.next = ll2.head.next.next
print(ll2.detect_cycle())   # True
````

### Problem: Detect a Cycle in Linked List

**Pattern:**  
Slow and fast pointer — Floyd's algorithm  

**Key Trick:**  
- No cycle → fast hits `None`  
- Cycle exists → fast laps slow, they meet  

**Same pattern as Find Middle — different condition:**  
- **Find Middle:** when fast **STOPS**  
- **Detect Cycle:** when fast **MEETS** slow  

**Stop condition:**  
`fast != None and fast.next != None`  

**Return:**  
- `True` if `slow == fast` inside loop  
- `False` if loop ends naturally  

**Time Complexity:** `O(n)`  
**Space Complexity:** `O(1)`
