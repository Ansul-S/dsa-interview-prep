## Interview Problem 1 — Reverse a Linked List

### The Problem
Given a linked list, reverse it **in place**.

**Input:**  
`head → [1] → [2] → [3] → [4] → [5] → None`

**Output:**  
`head → [5] → [4] → [3] → [2] → [1] → None`

```python
class Node():
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
            return

        current = self.head
        while current.next != None:
            current = current.next

        current.next = new_node


    def reverse(self):
        prev = None
        current = self.head

        while current != None:
            next = current.next
            current.next = prev
            prev = current
            current = next
        self.head = prev

    def traverse(self):
        current = self.head
        while current != None:
            print(current.data)
            current = current.next

ll = LinkedList()
ll.append(1)
ll.append(2)
ll.append(3)
ll.append(4)
ll.append(5)

print("Before:")
ll.traverse()

ll.reverse()

print("After:")
ll.traverse()
```
