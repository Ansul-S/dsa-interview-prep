````python
# Singly Linked List

class Node:
    def __init__(self, info, next=None):
        self.data = info
        self.next = next

class SinglyLinkedList:
    def __init__(self, head=None):
        self.head = head

    def insertAtEnd(self, value):
        temp = Node(value)

        if self.head is not None:
            t1 = self.head
            while t1.next is not None:
                t1 = t1.next
            t1.next = temp
        else:
            self.head = temp

    def printLL(self):
        t1 = self.head
        while t1 is not None:
            print(t1.data)
            t1 = t1.next

# usage
obj = SinglyLinkedList()
obj.insertAtEnd(10)
obj.insertAtEnd(20)
obj.insertAtEnd(30)
obj.printLL()
````

### Concept: LinkedList Class — Append + Traverse

**Key Trick:**
- `append` has two situations:
  1. **Empty list** → new node becomes `head`
  2. **Non-empty list** → travel to last node  
     → `current.next = new_node`

**Last Node Check:**
- `while current.next != None` → keep traveling  
- when `current.next == None` → you're at the end  

**Time Complexity:**
- Append: `O(n)`
- Traverse: `O(n)`

**Space Complexity:**
- `O(1)`

````python
class Node():
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
            return
        current = self.head
        while current.next != None:
            current = current.next

        current.next = new_node

    def prepend(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def traverse(self):
        current = self.head
        while current != None:
            print(current.data)
            current = current.next

ll = LinkedList()
ll.append(3)
ll.append(7)
ll.append(1)
ll.prepend(9)   # should appear at the front
ll.prepend(0)   # should appear before 9
ll.traverse()
````

### Concept: Prepend — Add to Beginning

**Steps:**
1. `new_node.next = self.head`  ← connect first  
2. `self.head = new_node`       ← redirect head  

**Key Trick:**
- Order matters — **connect THEN redirect**
- Reverse order = node points to itself   

**Time Complexity:**
- `O(1)` — no travelling needed  

**Space Complexity:**
- `O(1)`

**Compare:**
- `append` = `O(n)`  
- `prepend` = `O(1)`

````python
class Node():
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
            return
        current = self.head
        while current.next != None:
            current = current.next

        current.next = new_node

    def prepend(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def delete(self, data):
        if self.head is None:
            return
        if self.head.data == data:
            self.head = self.head.next
            return
        current = self.head
        while current.next != None:
            if current.next.data == data:
                current.next = current.next.next
                return
            current = current.next
        print(f'{data} not found')
    def traverse(self):
        current = self.head
        while current != None:
            print(current.data)
            current = current.next

ll = LinkedList()
ll.append(3)
ll.append(7)
ll.append(1)
ll.append(9)
ll.traverse()

print("---")
ll.delete(7)    # delete middle node
ll.traverse()

print("---")
ll.delete(3)    # delete head
ll.traverse()

print("---")
ll.delete(5)    # not found

````

###  Linked List Deletion — Core Concept

**Concept:**  
Delete a node

**Key Trick:**  
- Look **ONE step ahead**  
- Stand at the **previous node**, not the target node  
- `current.next.data == target` → found it  
- `current.next = current.next.next` → skip it  

---

###  Three Situations

1. **Empty list** → return  
2. **Delete head** → `self.head = self.head.next`  
3. **Middle node** → `current.next = current.next.next`  

---

### ⏱ Complexity

- **Time:** `O(n)`  
- **Space:** `O(1)`

````python
class Node():
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
            return
        current = self.head
        while current.next != None:
            current = current.next

        current.next = new_node

    def prepend(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def delete(self, data):
        if self.head is None:
            return
        if self.head.data == data:
            self.head = self.head.next
            return
        current = self.head
        while current.next != None:
            if current.next.data == data:
                current.next = current.next.next
                return
            current = current.next
        print(f'{data} not found')
    def search(self, data):
        current = self.head
        while current != None:
            if current.data == data:
                return True
            current = current.next
        return False


ll = LinkedList()
ll.append(3)
ll.append(7)
ll.append(1)
ll.append(9)

print(ll.search(7))   # True
print(ll.search(5))   # False
````

###  Linked List Search — Core Concept

**Concept:**  
Search a linked list  

**Pattern:**  
Traversal + condition check  

**Key Trick:**  
- `return True` inside loop → found  
- `return False` after loop → never found  

---

### ⏱ Complexity

- **Time:** `O(n)`  
- **Space:** `O(1)`
