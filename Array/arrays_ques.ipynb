### Q1: Find Maximum Element

Given:

arr = [4, 7, 1, 9, 2]

Find the largest element in the array.

Constraints:
- Solve in O(n)

Questions:
1. Brute force approach?
2. Optimal approach?

```python
# My Answer:
# Brute force
arr = [4,7,1,9,2]
print(max(arr))

# Optimal
arr = [4,7,1,9,2]
max_val = arr[0]
for x in arr:
    if x > max_val:
        max_val = x
print(max_val)
```

```python
# Correction answer 1
# My brute force is wrong in interview it is built in function
for i in range(len(arr)):
    is_largest = True

    for j in range(len(arr)):
        if arr[j] > arr[i]:
            is_largest = False
            break

    if is_largest:
        print(arr[i])


# Better Approch for optimal
max_val = arr[0]

for x in arr[1:]:
    if x > max_val:
        max_val = x
print(max_val)
```

# Q2: Second Largest

Given:

arr = [10, 5, 8, 20, 15]

Find the second largest element.

Constraint:
- Try solving in one pass.

Questions:
1. Brute force?
2. One-pass solution?

```python
# Brute Force
arr = [10, 5, 8, 20, 15]

# find largest
first = arr[0]
for x in arr:
    if x > first:
        first = x

# find second largest
second = float('-inf')
for x in arr:
    if x != first and x > second:
        second = x

print(second)
```

```python
# One-pass solution
arr = [10, 5, 8, 20, 15]

first = float('-inf')
second = float('-inf')

for x in arr:
    if x > first:
        second = first
        first = x
    elif x < first and x > second:
        second = x

print(second)
```

```python
# Find second largest in ONE PASS

arr = [10, 5, 8, 20, 15]

# We use -infinity so that ANY number in the array will be bigger initially
first = float('-inf')    # largest so far
second = float('-inf')   # second largest so far

for x in arr:
    # Case 1: Found a new largest number
    if x > first:
        second = first   # old largest becomes second largest
        first = x        # update largest

    # Case 2: Not largest, but can be second largest
    elif x < first and x > second:
        second = x

# After one full pass, 'second' holds the answer
print("Second largest:", second)
```
