# Arrays

## Types of Arrays

### 1. Homogeneous Arrays (Homo)
- Store **elements of the same data type**
- Example: **C Language Arrays**
- All elements must be uniform  
- Example:

```c
int arr[] = {1, 2, 3, 4};
```

---

### 2. Heterogeneous Arrays (Hetero)
- Can store **different data types**
- Common in **Python** (especially lists)

Example:

```python
arr = [10, "Hello", 3.14, True]
```

---

## Definition of an Array

An array is:

- A **collection of similar elements**  
- Stored in **contiguous memory locations**  
- Allows efficient access using **indexing**

> **Note:**  
> Traditional arrays usually contain similar data types,  
> but Python structures can also support mixed data types.

---

## Contiguous Memory Allocation

Arrays store elements in **continuous (adjacent) memory locations**.

Example:

| Index | Value | Memory Location |
|-------|-------|----------------|
| 0 | 10 | 1000 |
| 1 | 20 | 1004 |
| 2 | 30 | 1008 |

This enables:

- Faster access  
- Efficient traversal  
- Better memory management

---

## Accessing Arrays in Python

### 1. Using Python `array` Module

```python
import array as arr

a = arr.array('i', [10, 20, 30])

print(a[0])   # Access first element
```

---

### 2. Using NumPy

```python
import numpy as np

a = np.array([10, 20, 30])

print(a[1])   # Access second element
```

## inf logic
### Why do we use it here?

We need `second_max` to start at some value. The options are:

```python
second_max = None          # ❌ Python can't compare None with numbers
second_max = 0             # ❌ Breaks if all numbers are negative
second_max = arr[1]        # ❌ Might accidentally skip the real answer
second_max = float('-inf') # ✅ Always safe — any real number beats it
```

It's basically saying:

> "I haven't found a second largest yet, but I'm ready to accept anything."

---

## A real-life analogy

Imagine a high score leaderboard.

Before anyone plays, the **#2 slot says `−∞`**.

The first person who scores anything — even 1 point — immediately becomes #2.

That’s exactly what `float('-inf')` does here.

---

## The twin you'll also see

```python
float('inf')   # positive infinity — every number is smaller than it
```

You'll use this when finding minimum values:

```python
min_val = float('inf')   # same idea, opposite direction
```

---

## Bookmark this in your brain 🧠

- Finding **max** → initialize with `float('-inf')`
- Finding **min** → initialize with `float('inf')`

You'll use these two lines hundreds of times in your coding life. Now you know why.

# 🚀 Move Zeroes

## 🧬 The Biology Analogy First

Imagine you're looking at a cell sample under a microscope:

- **Healthy cells** → non-zero numbers  
- **Dead cells** → zeroes  

🎯 **Goal:** Push all dead cells to the right, keep healthy cells on the left — **in their original order**

**Example:**
Before: [0, 1, 0, 3, 12]
After: [1, 3, 12, 0, 0]

✅ The healthy cells **(1, 3, 12)** kept their order — we just removed the dead ones from between them.

---

## 🧠 The Two Pointer Idea — Plain English

Use **two pointers**, both starting from the left:

- **slow pointer** → marks *where the next healthy cell should go*
- **fast pointer** → scouts ahead looking for healthy cells
[0, 1, 0, 3, 12]
↑
slow
↑
fast


---

## ⚙️ How It Works

- `fast` sees **0** → dead cell → skip  
- `fast` sees **1** → healthy → swap with `slow`, move `slow`  
- `fast` sees **0** → skip  
- `fast` sees **3** → swap, move `slow`  
- `fast` sees **12** → swap, move `slow`  

🎯 **Final Result:**

[1, 3, 12, 0, 0] ✅

---

## 🔍 Step-by-Step Visualization
arr = [0, 1, 0, 3, 12]

slow = 0, fast = 0

fast=0 → arr[0]=0 → dead → skip → fast=1

fast=1 → arr[1]=1 → healthy → swap(arr[slow], arr[fast])
[1, 0, 0, 3, 12]
slow=1, fast=2

fast=2 → arr[2]=0 → dead → skip → fast=3

fast=3 → arr[3]=3 → healthy → swap(arr[slow], arr[fast])
[1, 3, 0, 0, 12]
slow=2, fast=4

fast=4 → arr[4]=12 → healthy → swap(arr[slow], arr[fast])
[1, 3, 12, 0, 0]
slow=3, fast=5

fast=5 → out of bounds → STOP ✅

---

## 📜 The Rule (Core Logic)
if fast pointer sees a NON-ZERO:
swap it with slow pointer position
move slow forward

always move fast forward

---

## ⚡ Key Insight

- Maintains **relative order** (stable)
- Done **in-place** (no extra space)
- Time complexity: **O(n)**

---

## 🧠 Mental Model

> "Slow builds the final answer. Fast finds the next useful element."

# 🚀 Remove Duplicates (Sorted Array)

## 🧠 Plain English First

Given a **sorted array**, remove duplicates so each element appears only once.

**Example:**
Input: [1, 1, 2, 3, 3, 4]
Output: [1, 2, 3, 4]

---

## 📌 Why "Sorted" Matters

Because duplicates are always **next to each other**:


[1, 1, 2, 3, 3, 4]
↑ duplicate sits beside original


❗ If it was unsorted:

[1, 3, 1, 2, 3]

Duplicates are scattered → much harder.

👉 Sorting gives you **structure = easy detection**

---

## 🔁 The Pattern — Same as Move Zeroes

Two pointers again:

- **slow pointer** → where next *unique* element should go  
- **fast pointer** → scans array  

💡 Only change is the definition of "valid":

- Move Zeroes → valid = **non-zero**
- Remove Duplicates → valid = **different from previous**

---

## 🔍 Step-by-Step Trace


arr = [1, 1, 2, 3, 3, 4]

slow = 0

fast=0 → arr[0]=1 → first element → always keep

fast=1 → arr[1]=1 → same as arr[slow]=1 → duplicate → skip

fast=2 → arr[2]=2 → different from arr[slow]=1 → keep
slow=1 → arr[1]=2

fast=3 → arr[3]=3 → different from arr[slow]=2 → keep
slow=2 → arr[2]=3

fast=4 → arr[4]=3 → same as arr[slow]=3 → duplicate → skip

fast=5 → arr[5]=4 → different from arr[slow]=3 → keep
slow=3 → arr[3]=4


---

## ✅ Final State


[1, 2, 3, 4, ?, ?]


✔ First `slow + 1 = 4` elements are unique  
✔ Ignore the rest

---

## 📜 The Rule (Core Logic)


if arr[fast] != arr[slow]:
move slow forward
place arr[fast] at arr[slow]

always move fast forward


---

## ⚡ Key Insights

- Works because array is **sorted**
- Done **in-place** (no extra space)
- Time complexity: **O(n)**

---

## 🧠 Mental Model

> "Slow builds the unique list. Fast finds the next new value."

---

## 🔥 Pattern Connection

Both problems follow the same skeleton:

| Problem              | Condition (valid element)       |
|---------------------|--------------------------------|
| Move Zeroes         | `arr[fast] != 0`               |
| Remove Duplicates   | `arr[fast] != arr[slow]`       |

👉 Same engine, different rule.
